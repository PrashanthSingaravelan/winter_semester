```python
ip = input("Enter the IP in binary format : ")
id1, id2, id3, id4 = ip.split(".")

if len(id1) == len(id2) == len(id3) == len(id4) == 8 :
    
    id1 = int("0b" + id1, 2)
    id2 = int("0b" + id2, 2)
    id3 = int("0b" + id3, 2)
    id4 = int("0b" + id4, 2)
    
    if id1 in range(0,128):
        print("\nCLASS A IP ADDRESS")
        print("Host id    :",id1)
        print(f"Network id : {id2}.{id3}.{id4}")
    
    if id1 in range(128,192):
        print("\nCLASS B IP ADDRESS")
        print(f"Host id    : {id1}.{id2}")
        print(f"Network id : {id1}.{id2}")
        
    if id1 in range(192,224):
        print("\nCLASS C IP ADDRESS")
        print(f"Host id    : {id1}.{id2}.{id3}")
        print("Network id :",int(id4,2))
        
    if id1 in range(224,240):
        print("\nCLASS D IP ADDRESS")
        print("No Host and Network ids")
        
    if id1 in range(240,256):
        print("\nCLASS E IP ADDRESS")
        print("No Host and Network ids")
    
else:
    print("\nOops...Invalid Address given !!!")

# 11111111.11111111.11111111.11111111
```
